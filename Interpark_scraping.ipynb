```python
from selenium import webdriver
```

```python
browser = webdriver.Chrome('./chromedriver.exe')
```

```python
browser.get('http://ticket.interpark.com/Contents/Ranking?smid1=s_menu&smid2=performance')
```

```python
from bs4 import BeautifulSoup
```

```python
html = browser.page_source
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
len(soup), type(soup)
```

```python
tags = soup.select('div.items')
tags
```

```python
tag = tags[0]
tag
```

```python
title = tag.select('a.prdName')
title
```

```python
contents =[]
for tag in tags:
    title = tag.select('a.prdName')
    duration = tag.select('a.prdDuration')
    musical = [title[0].text.strip(),duration[0].text.strip()]
    contents.append(musical)
contents
```

```python
import pandas as pd
```

```python
pd_data = pd.DataFrame(contents, columns=['title','Data Range'])
```

```python
pd_data.to_excel('./saves/interpark_scraping.xls')
```
